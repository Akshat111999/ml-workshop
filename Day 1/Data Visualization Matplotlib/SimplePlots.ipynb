# Matplotlib - pyplot

#### Import Library

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 4, 5, 6, 7, 8, 9, 10]
```

```python
plt.plot(x,y)
plt.show()
```

### Customizing the plot

```python
plt.plot(x, y)
plt.title('Customising the Plot')
plt.xlabel('Time in Years')
plt.xticks([])
plt.axis([0,20, 0, 10])

plt.ylabel('Some random Numbers')
```

```python
plt.scatter(x, y)
```
